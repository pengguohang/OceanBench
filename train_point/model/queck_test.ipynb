```python
import torch
import torch.nn as nn

embedding_date = nn.Embedding(5400, 16)

# 假设输入是日期索引 3 和 500
input_indices = torch.tensor([3, 500])

# 得到对应的嵌入向量
output = embedding_date(input_indices)
print(output)  # 输出是一个 [2, 16] 的张量，表示两个索引对应的 16 维嵌入
```

```python
x = torch.rand((16, 12, 24, 48, 6))
y = torch.rand((16, 12, 24, 48, 102))

input_shape = x.shape[1:]
target_shape = y.shape[1:]
downsample_scale = [1, 2, 2]

model = TransformerModel(input_shape,
                        target_shape,
                        base_units=64,
                        downsample_scale=downsample_scale,
                        pos_embed_type = "t+h+w",
                        z_init_method='nearest_interp',
                        block_units=None)

y1 = model(x)
print(y1.shape)
```
